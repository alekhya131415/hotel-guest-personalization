## Feature Engineering

### Guest Activity Affinity Score

```python
df["guest_activity_affinity"] = (
    df["spa_affinity_score"] +
    df["adventure_affinity_score"] +
    df["culture_affinity_score"] +
    df["dining_affinity_score"]
) / 4
```

```python
df["guest_activity_affinity"].describe()
```

* The combined activity affinity score summarizes overall guest engagement preferences and provides a holistic indicator of interest intensity.

### Event Popularity Index

```python
df["event_popularity_index"] = (
    df["event_popularity_score"] / 100
)

df["event_popularity_index"].describe()
```

* The normalized event popularity index standardizes event attractiveness and allows consistent comparison across features.

### Weather Comfort Rating

```python
df["weather_comfort_rating"] = (
    (1 - df["rain_probability"]/100) *
    (1 - abs(df["temperature"] - 25)/25)
)

df["weather_comfort_rating"].describe()
```

* The weather comfort rating quantifies environmental suitability for outdoor activities, enhancing context-aware recommendations.

### Preference Diversity Score

```python
df["preference_diversity"] = df[
    ["spa_affinity_score",
     "adventure_affinity_score",
     "culture_affinity_score",
     "dining_affinity_score"]
].std(axis=1)

df["preference_diversity"].describe()
```

* The preference diversity score captures variability in guest interests. Higher values indicate diverse preferences, enabling more flexible recommendations.

### Experience Match Score

```python
df["experience_match_score"] = (
    df["weather_activity_match"] +
    df["event_preference_alignment"]
) / 2

df["experience_match_score"].describe()
```

* The experience match score quantifies alignment between guest preferences and contextual conditions, improving recommendation precision.


#### Overall Summary:-
* New composite features were created to capture guest behavior, contextual conditions, and preference alignment.
* Aggregated affinity scores summarize guest interests, while normalized event and weather metrics enhance comparability.
* Additional diversity and match scores improve the system’s ability to personalize recommendations.
* These engineered features strengthen predictive modeling and support intelligent hospitality personalization.

##### Recommended experience for a guest

```python
import numpy as np

def recommend_experience(guest_index, top_n=5):
    similarities = similarity_matrix[guest_index]
    
    # Get most similar guests
    similar_guests = np.argsort(similarities)[::-1][1:top_n+1]
    
    # Get their recommended experiences
    recommendations = df_sample.iloc[similar_guests]["recommended_experience"]
    
    return recommendations.mode()[0]
```

##### Hybrid Recommendation Model

```python
gb_predictions = model.predict(X_test)
```

```python
# Hybrid Prediction Function
def hybrid_recommendation(index):
    
    # Gradient Boosting prediction
    gb_pred = model.predict([X_test.iloc[index]])[0]
    
    # Collaborative Filtering prediction
    cf_pred = recommend_experience(index)
    
    # Hybrid logic
    if gb_pred == cf_pred:
        return gb_pred
    else:
        return gb_pred
```

```python
# Generate Hybrid Predictions
import warnings
warnings.filterwarnings("ignore")
hybrid_predictions = []

for i in range(len(X_test)):
    pred = hybrid_recommendation(i)
    hybrid_predictions.append(pred)
```

```python
# Evaluate Hybrid Model
from sklearn.metrics import accuracy_score

hybrid_accuracy = accuracy_score(y_test, hybrid_predictions)

print("Hybrid Model Accuracy:", hybrid_accuracy)
```

* The hybrid recommendation model integrates Gradient Boosting predictions with collaborative filtering similarity.
* This approach combines behavioral learning and peer-based recommendations, improving personalization robustness.
* While performance gains are modest due to synthetic data limitations, the hybrid architecture demonstrates scalable recommendation design principles.

##### Evaluate across guest segments

```python
segment_accuracy = {}

for segment in df["travel_type"].unique():
    
    # Filter original dataframe first
    df_seg = df[df["travel_type"] == segment]
    
    # Recreate X and y from the filtered data
    X_seg = pd.get_dummies(
        df_seg.drop(["guest_id", "recommended_experience"], axis=1),
        drop_first=True
    )
    
    y_seg = df_seg["recommended_experience"]
    
    scores = cross_val_score(
        model,
        X_seg,
        y_seg,
        cv=3,
        scoring="accuracy"
    )
    
    segment_accuracy[segment] = scores.mean()

print(segment_accuracy)
```

```python
# Convert SHAP list output into usable array

# If shap_values is a list (multiclass)
if isinstance(shap_values, list):
    shap_selected = shap_values[0]   # choose one class to explain
else:
    shap_selected = shap_values.values
```

```python
# Fix Explain Guest Function
import numpy as np

def explain_guest(index):
    
    shap_vals = shap_selected[index]
    feature_names = X_test.columns
    
    # Top 3 strongest drivers
    top_idx = np.argsort(np.abs(shap_vals))[-3:]
    
    top_features = feature_names[top_idx]
    
    return top_features
```

```python
# Fix Explain Generator
def generate_explanation(index):
    
    experience = best_model.predict(X_test.iloc[[index]])[0]
    drivers = explain_guest(index)
    
    explanation = (
        f"{experience} recommended due to strong influence of "
        f"{drivers[0]}, {drivers[1]}, and {drivers[2]}."
    )
    
    return explanation
```
